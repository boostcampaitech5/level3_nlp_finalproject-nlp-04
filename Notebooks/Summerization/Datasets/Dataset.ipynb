```python
import json
import pandas as pd
from tqdm import tqdm
from typing import List
from datasets import Dataset
```

```python
# file_json = open("dataset/papaer/논문요약_0206_0.json", "w")
# paper_file = json.dump(json_file)
```

```python
def json_to_dict(file_path: str) -> List:
    list_paper = []
    
    # "dataset/papaer/논문요약_0206_0.json"
    # file_json = open(file_path, "w")
    # paper_json = json.load(file_json)
    with open(file_path, 'r') as f:
        paper_json = json.load(f)

    for idx, paper in enumerate(tqdm(paper_json["data"])):
        # doc_type = paper["doc_type"]
        doc_id = paper["doc_id"]
        title = paper["title"]
        ipc = paper["ipc"]

        # 전체요약 구분
        summary_entire_ori = paper["summary_entire"][0]["orginal_text"]
        summary_entire_sum = paper["summary_entire"][0]["summary_text"]

        # 섹션요약 구분
        summary_section_ori = paper["summary_section"][0]["orginal_text"]
        summary_section_sum = paper["summary_section"][0]["summary_text"]

        dict_entire = {"doc_id": doc_id, \
                       "title": title, "ipc": ipc, \
                       "all": summary_entire_ori, "sum": summary_entire_sum}
        dict_section = {"doc_id": doc_id + "_1", \
                       "title": title, "ipc": ipc, \
                       "all": summary_section_ori, "sum": summary_section_sum}

        list_paper.append(dict_entire)
        list_paper.append(dict_section)

    return list_paper
```

```python
list_path = ["dataset/paper/논문요약_0206_0.json", "dataset/paper/논문요약_0206_1.json",\
            "dataset/paper/논문요약_0206_2.json", "dataset/paper/논문요약_0220_0.json",\
            "dataset/paper/논문요약_0225_5_1.json", "dataset/paper/논문요약_0225_7_0.json"]
```

```python
list_paper = []

for path in list_path:
    list_paper = list_paper + json_to_dict(path)
```

```python
df_paper = pd.DataFrame(list_paper)
print(len(df_paper))
df_paper.sample(20)
```

```python
dataset = Dataset.from_pandas(df_paper)
dataset = dataset.train_test_split(test_size=0.002, seed=980217)
dataset.save_to_disk("./dataset/paper/paper_dataset")
dataset
```
